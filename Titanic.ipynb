```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('tested.csv')

# Fill missing values to avoid plotting issues
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

# Subset with numeric columns only
numeric_cols = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']

# Pairplot
sns.pairplot(df[numeric_cols], hue='Survived', palette='coolwarm')
plt.suptitle("Pairplot of Titanic Data by Survival", y=1.02)
plt.show
```

Histograms (Distribution)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for Age and Fare
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Age'].fillna(df['Age'].median()), bins=30, kde=True, color='skyblue')
plt.title("Age Distribution")

plt.subplot(1, 2, 2)
sns.histplot(df['Fare'].fillna(df['Fare'].median()), bins=30, kde=True, color='salmon')
plt.title("Fare Distribution")

plt.tight_layout()
plt.show()
```

Boxplots (Outliers by Survival)

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")

plt.subplot(1, 2, 2)
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare Distribution by Survival")

plt.tight_layout()
plt.show()
```

Scatterplot (Age vs Fare)

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='cool')
plt.title("Age vs Fare by Survival")
plt.show()
```
